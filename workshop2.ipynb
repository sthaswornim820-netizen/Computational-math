```python
def unique_list(lst):
    list2 = []

    for i in lst:
        if i not in list2:
            list2.append(i)

    return list2

list1 = [1,2,3,3,3,3,4,5]
print(unique_list(list1))
```

```python
def is_prime(n):
    # Check if number is less than or equal to 1
    if n <= 1:
        return False

    # Loop from 2 to n-1
    for i in range(2, n):
        # Check if n is divisible by i
        if n % i == 0:
            return False  # Not prime if divisible

    return True  # Prime if divisible

num = 4

# calling function and printing result
if is_prime(num):
    print("Prime number")
else:
    print("Not a prime number")
```

```python
# Function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True


# Print twin primes less than 1000
for i in range(3, 1000):
    # Check if both i and i+2 are prime
    if is_prime(i) and is_prime(i + 2):
        print(i, "and", i + 2)
```

```python
def sumOfDivisors(n):
    total = 0  # to store sum of divisors

    # loop from 1 to n-1
    for i in range(1, n):
        if n % i == 0:   # check if i is a divisor
            total += i   # add divisor to total

    return total

num = 6
print(sumOfDivisors(num))
```

```python
# Function to find sum of divisors
def sumOfDivisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total


# Find next perfect number after 6
num = 7

while True:
    if sumOfDivisors(num) == num:
        print("Next perfect number is:", num)
        break
    num += 1
```

```python
# Given list
marks = [20, 30, 35, 55, 70, 80, 90]

# Find mean
sum = 0
for i in marks:
    sum = sum + i   # add each number

mean = sum / len(marks)


# Find Median
n = len(marks)
median = marks[n // 2]

print("Mean =", mean)
print("Median =", median)
```

```python
def decimal_to_binary(n):
    binary = ""

    while n > 0:
        remainder = n % 2      # get remainder (0 or 1)
        binary = str(remainder) + binary  # add to front
        n = n // 2             # divide number by 2

    return binary

num = 10
print("Binary =", decimal_to_binary(num))
```

```python
def prodDigits(n):
    product = 1

    # loop until number becomes 0
    while n > 0:
        digit = n % 10      # get last digit
        product = product * digit
        n = n // 10         # remove last digit

    return product

num = 456
print("Product of digits =", prodDigits(num))
```
