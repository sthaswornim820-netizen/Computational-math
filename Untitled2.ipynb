```python
import numpy as np
import matplotlib.pyplot as plt
import random
#We will start with a simple problem. Consider the following linear systems: a. x+3y = 5 𝑥−2𝑦=−5 𝑥+𝑦=3

x = np.linspace(-5,5,100)

y1 = (5-x)/3
y2 = (5+x)/2
y3 = 3-x

fig, ax = plt.subplots()

ax.plot(x,y1, color='r')
ax.plot(x,y2, color='g')
ax.plot(x,y3, color='b')

ax.text(0.5,0.6, '$x+3y = 5$', color='r')
ax.text(-3,0.5,'$x-2y = -5$', color='g')
ax.text(1.2,2, '$x+y = 3$', color='b')

ax.set_xlim(-4,4)
ax.set_ylim(-2,6)

ax.axvline(color='k',linewidth = 1)
ax.axhline(color='k',linewidth = 1)

ax.set_xticks(list(range(-4,5)))
ax.set_aspect('equal')
ax.grid(True,ls=':')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x = np.linspace(-5,5,100)

y1 = (3-2*x)/3
y2 = (6-4*x)/6
y3 = (1-3*x)/(-2)

fig, ax = plt.subplots()

ax.plot(x,y1, color='r')
ax.plot(x,y2, color='g')
ax.plot(x,y3, color='b')

ax.text(0.5,0.6, '$2x + 3y = 3$', color='r')
ax.text(-3,0.5,'$4𝑥+6𝑦=6$', color='g')
ax.text(1.2,2, '$3𝑥−2𝑦=1$', color='b')

ax.set_xlim(-4,4)
ax.set_ylim(-2,6)

ax.axvline(color='k',linewidth = 1)
ax.axhline(color='k',linewidth = 1)

ax.set_xticks(list(range(-4,5)))
ax.set_aspect('equal')
ax.grid(True,ls=':')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
