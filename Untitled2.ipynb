```python
#Task 6

#Create three vectors in 3D space and plot them using Matplotlib.

origin = np.array([0,0])

V = np.array([1,1])

W = np.array([-2.0,-1])

fig, ax = plt.subplots()

# Add vectors V and W to the plot

ax.quiver(origin[0], origin[1], V[0], V[1], angles='xy', scale_units='xy', scale=1, color='r')

ax.quiver(origin[0], origin[1], W[0], W[1], angles='xy', scale_units='xy', scale=1, color='b')

ax.text(origin[0]-0.1, origin[1]+0.1, "O")

ax.text(V[0], V[1], "v")

ax.text(W[0]-0.125, W[1]-0.1, "w")
ax.set_xlim([-3, 3])

ax.set_ylim([-3, 3])

ax.set_xlabel('X')

ax.set_ylabel('Y')

plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

U = np.array([1, 0.25, 0])
V = np.array([0, 1, 0.25])
W = np.array([0.25, 0, 1])
origin = np.array([0, 0, 0])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(origin[0], origin[1], origin[2], V[0], V[1], V[2], color='r', label="v")
ax.quiver(origin[0], origin[1], origin[2], W[0], W[1], W[2], color='b', label="w")
ax.quiver(origin[0], origin[1], origin[2], U[0], U[1], U[2], color='y', label="u")

ax.set_xlim([0, 1.5])
ax.set_ylim([0, 1.5])
ax.set_zlim([0, 1.5])
ax.legend()

plt.show()
```

```python
#Write a function to calculate the direction cosines of the vector. For example, let the vector be O = (1.5, 1.5, 2.0).

O = [1.5,1.5,2.0]

O_Direction = []

Magnitude = (O[0]**2+O[1]**2+O[2]**2)**(1/2)

for i in O:
    O_Direction.append((i/Magnitude))

print(O_Direction)
```
