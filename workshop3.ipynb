```python
import matplotlib.pyplot as plt

cars = ["Red", "Silver", "Blue", "White", "Green"]
values = [123, 230, 78, 193, 12]

plt.bar(cars, values)

plt.title("Car Colors")
plt.xlabel("Color")
plt.ylabel("Number seen")

plt.show()
```

```python
import matplotlib.pyplot as plt

cars = ["Red", "Silver", "Blue", "White", "Green"]
values = [123, 230, 78, 193, 12]

plt.bar(cars, values, width=0.3)

plt.title("Car Colors")
plt.xlabel("Color")
plt.ylabel("Number seen")

plt.show()
```

```python
import matplotlib.pyplot as plt

cars = ["Red", "Silver", "Blue", "White", "Green"]
values = [123, 230, 78, 193, 12]

myexplode = [0.3, 0, 0, 0, 0]

plt.pie(values, labels=cars, explode=myexplode)
plt.title("Car Colors")

plt.show()
```

```python
import matplotlib.pyplot as plt

cars = ["Red", "Silver", "Blue", "White", "Green"]
values = [123, 230, 78, 193, 12]

explode = [0.3, 0, 0, 0, 0]

plt.pie(values, labels=cars, explode=explode, shadow=True)
plt.title("Car Colors")

plt.show()
```

```python
import matplotlib.pyplot as plt

cars = ["Red", "Silver", "Blue", "White", "Green"]
values = [123, 230, 78, 193, 12]

plt.pie(values, labels=cars)
plt.legend(title="Car Colours:")
plt.title("Car Colors")

plt.show()
```

```python
import matplotlib.pyplot as plt

cars = ["Red", "Silver", "Blue", "White", "Green"]
values = [123, 230, 78, 193, 12]

mycolors = ["magenta", "green", "cyan", "red", "yellow"]

plt.pie(values, labels=cars, colors=mycolors)
plt.title("Car Colors")

plt.show()
```
