```python
def sumDigits(n):
    sum = 0
    for i in range(1, n+1):
        sum += i
    print(sum)

sumDigits(6)
```

```python

def squaredSum(numbers):
    sum = 0
    for num in numbers:
        sum += num*num
    return sum

print(squaredSum([2,3,4]))
```

```python

def sumDigits(num):
    sum = 0
    while num != 0:
        remainder = num % 10
        sum += remainder
        num = num//10
    print(sum)

sumDigits(234)
```

```python
def myPow(x, n):
    i = 0
    ans = 1;
    while i < n:
        ans = ans * x
        i += 1
    return ans


print(myPow(3,2))
```

```python
# Declairing function

def fib(n):
    a = 0
    b = 1
    print(a)
    print(b)
    for i in range(0, n):
        c = a + b
        print(c)

        a = b
        b = c


# Prompting the user for a number
n = int(input("Enter any number: "))

# Function call
fib(n)
```

```python
def number_group(number):
  if number > 0:
    return "Positive"
  elif number == 0:
    return "Zero"
  else:
    return "Negative"
print(number_group(-5));
```

```python
def is_Even(number):
  if (number % 2 == 0):
    return True
  return False
print(is_Even(4));
```

```python
def order_numbers(number1, number2):
  if number2 > number1:
    return number1, number2;
  else:
    return number2, number1;
order_numbers(1, 5);
print(order_numbers(1, 5));
```
